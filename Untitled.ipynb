```python
import pandas as pd
```

```python
df = pd.read_parquet("final_flower_america_df.parquet")
df = df.dropna(subset = ['plant_type', 'product_type'])
df['how_many_flowers'] = df['flower_type'].apply(len)
df = df.drop(columns = ['Product Type', 'Store Type','Store Quality', 'filename', 'flowers_in_name', 'name_normalized', 'plant_type'])
df.columns = ['产品', '描述', '价格', '独立站', '产品链接', '图片链接', '品类', '颜色', '叶', '类型', '载具', '花种类数量']
df.head()
```

```python
product_type_map = {
    "single stem": "single",
}
```

```python
df['类型'] = df['类型'].replace(product_type_map)
```

```python
df.to_parquet("america_flower_df_chinese.parquet")
```
